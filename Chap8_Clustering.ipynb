##K-Means


Gennerate Data

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=6)
```

EXPLORE DATA

```python
X.shape
```

```python
from matplotlib import pyplot as plt 
plt.scatter(X[:,0],X[:,1])
```

###Clustering

import

```python
from sklearn.cluster import KMeans
```

Define

```python
kmeans = KMeans(n_clusters=2, random_state=6) #กำหนดคลัสเตอร์เเละเเรนด้อมเสตท
```

Train-Predict

```python
y_clust = kmeans.fit_predict(X)
```

```python
y_clust #ผลคือดาต้าตัวที่1อยู่กลุ่มที่1,ดาต้าตัวที่2อยุ่กลุ่มที่ศูนย์
```

```python
plt.scatter(X[:,0],X[:,1],c=y_clust) 
```

###Clusters

```python
kmeans = KMeans(n_clusters=3, random_state=6) #define
y_clust = kmeans.fit_predict(X)#fit-predict
plt.scatter(X[:,0],X[:,1],c=y_clust)#plot
```

```python
kmeans = KMeans(n_clusters=4, random_state=14, init= 'random') #define
y_clust = kmeans.fit_predict(X)#fit-predict
plt.scatter(X[:,0],X[:,1],c=y_clust)#plot
```

```python
kmeans = KMeans(n_clusters=5, random_state=6, init= 'random') #define
y_clust = kmeans.fit_predict(X)#fit-predict
plt.scatter(X[:,0],X[:,1],c=y_clust)#plot
```

###Example Application(Color Quanization)

```python
import cv2
import os
```

```python
img = cv2.imread('/content/Siberian-Husky.png')
```

```python
img.shape
```

```python
RGB_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

###นับจำนวนสี

```python
import numpy as  np
```

###จัดกลุ่มสีให้เหลือ 16 สี

```python
np.unique(np.reshape(RGB_img,(RGB_img.shape[0]*RGB_img.shape[1],3)),axis=0).shape
```

###แปลงข้อมูลให้อยู่ในรูป row-column

```python
myquantize = KMeans(n_clusters=16,random_state=6)
```

```python
img_row = np.reshape(RGB_img,(RGB_img.shape[0]*RGB_img.shape[1],3))
img_row.shape
```

```python
final_color = myquantize.fit_predict(img_row)
```

###ใช้ centroid เป็นตัวแทนสี

```python
centroid = myquantize.cluster_centers_
```

```python
centroid
```

```python
new_img = np.zeros(RGB_img.shape)
for i in range(RGB_img.shape[0]):
    for j in range(RGB_img.shape[1]):
        new_img[i,j,:] = centroid[myquantize.predict([RGB_img[i,j,:]]),:]
```

```python
plt.imshow(new_img/255)
```

```python
plt.imsave('img.jpg',new_img/255)
```
