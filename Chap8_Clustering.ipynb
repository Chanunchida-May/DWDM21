##K-Means


Gennerate Data

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=6)
```

EXPLORE DATA

```python
X.shape
```

```python
from matplotlib import pyplot as plt 
plt.scatter(X[:,0],X[:,1])
```

###Clustering

import

```python
from sklearn.cluster import KMeans
```

Define

```python
kmeans = KMeans(n_clusters=2, random_state=6) #กำหนดคลัสเตอร์เเละเเรนด้อมเสตท
```

Train-Predict

```python
y_clust = kmeans.fit_predict(X)
```

```python
y_clust #ผลคือดาต้าตัวที่1อยู่กลุ่มที่1,ดาต้าตัวที่2อยุ่กลุ่มที่ศูนย์
```

```python
plt.scatter(X[:,0],X[:,1],c=y_clust) 
```

###Clusters

```python
kmeans = KMeans(n_clusters=3, random_state=6) #define
y_clust = kmeans.fit_predict(X)#fit-predict
plt.scatter(X[:,0],X[:,1],c=y_clust)#plot
```

```python
kmeans = KMeans(n_clusters=4, random_state=14, init= 'random') #define
y_clust = kmeans.fit_predict(X)#fit-predict
plt.scatter(X[:,0],X[:,1],c=y_clust)#plot
```

```python
kmeans = KMeans(n_clusters=5, random_state=6, init= 'random') #define
y_clust = kmeans.fit_predict(X)#fit-predict
plt.scatter(X[:,0],X[:,1],c=y_clust)#plot
```
